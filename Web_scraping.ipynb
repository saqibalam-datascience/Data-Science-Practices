```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
mobile_name = []
rating = []
number_of_ratings = []
price = []
storage = []
display = []
battery = []
camera = []
processor = []


webpage = requests.get(f'https://www.flipkart.com/search?q=samsung+phones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1').text
soup = BeautifulSoup(webpage)
container = soup.find_all('div',class_ = '_2kHMtA')

for mobile in container:
    mobile_name.append(mobile.find('div',class_ = '_4rR01T').text)
    rating.append(mobile.find('div',class_ = '_3LWZlK').text)
    number_of_ratings.append(mobile.find('span',class_ = '_2_R_DZ').text)
    price.append(mobile.find('div', class_='_30jeq3').text)

        
    # Find_all used because there are multiple attributes in a single tag, so fetching it by indexes under a single tag
    storage.append(mobile.find_all('li', class_='rgWa7D')[0].text)
    display.append(mobile.find_all('li', class_='rgWa7D')[1].text)
    camera.append(mobile.find_all('li', class_='rgWa7D')[2].text)
    battery.append(mobile.find_all('li', class_='rgWa7D')[3].text)
    processor.append(mobile.find_all('li', class_='rgWa7D')[4].text)

df = pd.DataFrame({'mobile_name' : mobile_name,
                'rating' : rating,
                'number_of_ratings' : number_of_ratings,
                'price' : price,
                'storage' : storage,
                'display' : display,
                'camera' : camera,
                'battery' : battery,
                'processor' : processor})
```

```python
# As we can see that there is warranty written in the processor column, now fixing it.
df.processor = np.where(df['processor'].str.contains('year', case=False), np.nan, df['processor'])
```
